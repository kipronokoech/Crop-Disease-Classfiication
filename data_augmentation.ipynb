```
def image_generator(path,num_images):
    '''
    Arguments:
    path - path to the images to the augmented
    num_images - number of images to be generated per one image
    
    '''
    for Dir, SubDir, filesList in os.walk(path):
        for index,files in enumerate(filesList):
            #read the image
            image=cv.imread(os.path.join(Dir,files))
            image=cv.cvtColor(image,cv.COLOR_BGR2RGB)
            image = np.expand_dims(image, axis=0)
            #define augmentation arguments
            aug = ImageDataGenerator(zoom_range=0.15,rotation_range=30,
                    width_shift_range=0.2,
                    height_shift_range=0.2,
                    shear_range=0.1,
                    horizontal_flip=True,
                    fill_mode="nearest")
            print(index+1,os.path.join(Dir,files))
            #define where the new images will the saved to, the prefix
            #and the extension
            imageGen = aug.flow(image, batch_size=1,save_to_dir=Dir, save_prefix="imageaug", save_format="jpg")
            # loop over examples from our image data augmentation generator
            total=0
            for i,image in enumerate(imageGen):
                total += 1
                if total == num_images:
                    break
#path to the images to be augmented
#with this path we will augment images of class: healthy wheat
path_to_images = os.path.join(os.getcwd(),"./train/healthy_wheat")
#number of augmated images per one image in the path
num = 3 
#uncomment the line below to perform data augmentation using the above function

#image_generator(path_to_images,num)
```
